```python
import numpy as np
```

```python
# time and signal
time= np.linspace(0,1,500)
ecg_signal= np.sin(2*np.pi * 5 * time)
noise= np.random.normal(0,0.1,500)
ecg_signal= ecg_signal + noise

# basic statistics
avg_signal= np.mean(ecg_signal)
max_signal= np.max(ecg_signal)
min_signal= np.min(ecg_signal)

print("Average ECG Signal: {:0.2f} mV".format(avg_signal))
print("Maximum ECG Signal: {:0.2f} mV".format(max_signal))
print("Minimum ECG Signal: {:0.2f} mV".format(min_signal))

# abnormal points
high_threshold= 1.2  #mV
low_threshold= -0.3  #mV

high_points= ecg_signal > high_threshold
low_points= ecg_signal < low_threshold

print("High abnormal points (> {:.1f} mV): {}".format(high_threshold, np.sum(high_points)))
print("Low abnormal points (< {:.1f} mV): {}".format(low_threshold, np.sum(low_points)))
print("Total abnormal points: {}".format(np.sum(high_points | low_points)))
```
