```python
import numpy as np
```

```python
# Simulate hemoglobin levels
patients=1000
hemoglobin=np.random.normal(14,1.5,patients)

# Categorize values
low= hemoglobin < 12
high= hemoglobin > 17

# Statistics and Display results
print("☆  BLOOD TEST SIMULATION RESULTS  ☆")

avg_hemo=np.mean(hemoglobin)
print("Average hemoglobin level: " , round(avg_hemo,2))

abnormal_count= np.sum(low|high)
print("Abnormal number of patients: " , abnormal_count)

abnormal_percentage = (abnormal_count/patients)*100
print("Abnormal percentage: " , round(abnormal_percentage,2),"%")

min_hemo=np.min(hemoglobin)
max_hemo=np.max(hemoglobin)
std_hemo=np.std(hemoglobin)
print("Minimum hemoglobin:", round(min_hemo, 2))
print("Maximum hemoglobin:", round(max_hemo, 2))
print("Standard deviation:", round(std_hemo, 2))

print("Low Count:", np.sum(low),"/", "Normal Count:", np.sum(~low & ~high),"/", "High Count:", np.sum(high))
print("\n------------------------------------------\n")

# user input
user_value=float(input("Enter your hemoglobin level to check your status: "))

# evaluate user input
if user_value <12:
    print("Your result: LOW - AT RISK")
elif user_value >17:
    print("Your result: HIGH - AT RISK")
else:
    print("Your result: NORMAL")
```
